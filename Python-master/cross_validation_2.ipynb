```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

X = iris.data
y = iris.target

print(X.shape)

plt.scatter(X[:,0], X[:,1], c=y, alpha=0.8)
```

```python
# train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.title("Train test split")
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, alpha=0.8)
plt.subplot(1, 2, 2)
plt.scatter(X_test[:,0], X_test[:,1], c=y_test, alpha=0.8)
plt.title("Test Set split")
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=6)

model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(KNeighborsClassifier(4), X_train, y_train, cv=5, scoring="accuracy").mean()
```

```python
val_score = []
for k in range(1, 50):
    score = cross_val_score(KNeighborsClassifier(k), X_train, y_train, cv=5).mean()
    val_score.append(score)

plt.plot(val_score)
```

```python
from sklearn.model_selection import validation_curve
```

```python
model = KNeighborsClassifier()
k = np.arange(1, 50)

train_score, val_score = validation_curve(
        estimator=model,
        X=X_train,
        y=y_train,
        param_name='n_neighbors',
        param_range=k,
        cv=5
    )
```

```python
plt.plot(k, val_score.mean(axis=1))
```

```python
plt.plot(k, val_score.mean(axis=1), label='validation score')
plt.plot(k, train_score.mean(axis=1), label='train score')

plt.xlabel('n_neighbors')
plt.ylabel('score')
plt.legend()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parm_grid = {'n_neighbors': np.arange(1, 50), 'metric': ['euclidean', 'manhattan']}

grid = GridSearchCV(KNeighborsClassifier(), parm_grid, cv=5)

grid.fit(X_train, y_train)
```

```python
print("best parmaetres " , grid.best_params_)
print("best score ", grid.best_score_)
print("test score ", grid.score(X_test, y_test))
print("test score ", grid.score(X_train, y_train))  # overfitting
```

```python
model = grid.best_estimator_
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, model.predict(X_test))
```
